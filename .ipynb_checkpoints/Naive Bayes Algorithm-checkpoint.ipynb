## Check for co-relations

```python
def plot_corr(df, size =4):
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
```

## Check distribution of classes in Iris Dataset

```python
num_setosa = len(df.loc[df['class'] == 0])
num_versicolor = len(df.loc[df['class'] == 1])
num_virginica = len(df.loc[df['class'] == 2])
num_total = num_setosa + num_versicolor + num_virginica
print("Number of setosa flowers: {0} ({1:2.2f}%)".format (num_setosa, (num_setosa/(num_total)) * 100))
print("Number of versicolor flowers: {0} ({1:2.2f}%)".format (num_versicolor, (num_setosa/(num_total)) * 100))
print("Number of virginica flowers: {0} ({1:2.2f}%)".format (num_virginica, (num_setosa/(num_total)) * 100))
```

### Splitting the data
70 % for training, 30% for testing

```python
from sklearn.cross_validation import train_test_split
feature_col_names = ['sepal-length-in-cm','sepal-width-in-cm','petal-length-in-cm','petal-width-in-cm']
predicted_col_names = ['class']

x = df[feature_col_names].values
y = df[predicted_col_names].values
split_test_size = 0.30

x_train, x_test, y_train, y_test = train_test_split (x,y,test_size=split_test_size, random_state=42)
```

We check if desired split is obtained

```python
print ("{0:0.2f}% in training set".format ((len(x_train)/len(df.index))*100))
print ("{0:0.2f}% in test set".format ((len(x_test)/len(df.index))*100))
```

#### Verify predicted value was split correctly

```python
print("Original Data Set:")

num_setosa = len(df.loc[df['class'] == 0])
num_versicolor = len(df.loc[df['class'] == 1])
num_virginica = len(df.loc[df['class'] == 2])
num_total = num_setosa + num_versicolor + num_virginica
print("Number of setosa flowers: {0} ({1:2.2f}%)".format (num_setosa, (num_setosa/(num_total)) * 100))
print("Number of versicolor flowers: {0} ({1:2.2f}%)".format (num_versicolor, (num_setosa/(num_total)) * 100))
print("Number of virginica flowers: {0} ({1:2.2f}%)".format (num_virginica, (num_setosa/(num_total)) * 100))

print ("\nTraining Data Set:")
num_setosa_train = len(y_train[y_train[:] == 0])
num_versicolor_train = len(y_train[y_train[:] == 1])
num_virginica_train  = len(y_train[y_train[:] == 2])
num_total_train = num_setosa_train + num_versicolor_train + num_virginica_train
print("Number of setosa flowers: {0} ({1:2.2f}%)".format (num_setosa_train, (num_setosa_train/(num_total_train)) * 100))
print("Number of versicolor flowers: {0} ({1:2.2f}%)".format (num_versicolor_train, (num_setosa_train/(num_total_train)) * 100))
print("Number of virginica flowers: {0} ({1:2.2f}%)".format (num_virginica_train, (num_setosa_train/(num_total_train)) * 100))

print ("\nTesting Data Set:")
num_setosa_test = len(y_test[y_test[:] == 0])
num_versicolor_test = len(y_test[y_test[:] == 1])
num_virginica_test  = len(y_test[y_test[:] == 2])
num_total_test = num_setosa_test + num_versicolor_test + num_virginica_test
print("Number of setosa flowers: {0} ({1:2.2f}%)".format (num_setosa_test, (num_setosa_test/(num_total_test)) * 100))
print("Number of versicolor flowers: {0} ({1:2.2f}%)".format (num_versicolor_test, (num_setosa_test/(num_total_test)) * 100))
print("Number of virginica flowers: {0} ({1:2.2f}%)".format (num_virginica_test, (num_setosa_test/(num_total_test)) * 100))
```

## Training with Algorithm - Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

#create Gaussian Naive Bayes model object and train the data
nb_model = GaussianNB()

nb_model.fit(x_train, y_train.ravel())
```

## Performance on Training Data

```python
# predict values using training data
nb_predict_train = nb_model.predict(x_train)

# import performance metrics library
from sklearn import metrics

# calculating accuracy
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_train, nb_predict_train)))
print()
```

## Performance on Testing Data

```python
# predict values using testing data
nb_predict_test = nb_model.predict(x_test)

# import performance metrics library
from sklearn import metrics

# calculating accuracy
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, nb_predict_test)))
print()
```

#### Metrics

```python
print("Confusion Matrix")
# Note the use of labels for set 0=Iris Setosa, 1=Iris Versicolor, 2=Iris Virginica
print("{0}".format(metrics.confusion_matrix(y_test, nb_predict_test, labels = [0,1,2])))
print()

print("Classification Report")
print(metrics.classification_report(y_test, nb_predict_test, labels = [0,1,2]))
```
